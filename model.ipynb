```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('C:\\Users\\soura\\OneDrive\\Documents\\CS661_proj\\data\\weather_states.csv')
```

```python
# bin 'temp' column
bins_temp = [-9999, 25, 30, 35, 9999]
labels_temp = ['<25', '25 to 30', '30 to 35', '>35']
data['temp_class'] = pd.cut(data['temp'], bins=bins_temp, labels=labels_temp)

# bin 'stp' column
bins_stp = [-9999, 980, 990, 1000, 9999]
labels_stp = ['<980', '980 to 990', '990 to 1000', '>1000']
data['stp_class'] = pd.cut(data['stp'], bins=bins_stp, labels=labels_stp)

# bin 'hmdy' column
bins_hmdy = [-9999, 25, 50, 75, 9999]
labels_hmdy = ['<25', '25 to 50', '50 to 75', '>75']
data['hmdy_class'] = pd.cut(data['hmdy'], bins=bins_hmdy, labels=labels_hmdy)

# bin 'prcp' column
bins_prcp = [-9999, 0.1, 1, 10, 9999]
labels_prcp = ['<0.1', '0.1 to 1', '1 to 10', '>10']
data['prcp_class'] = pd.cut(data['prcp'], bins=bins_prcp, labels=labels_prcp)

# bin 'wdct' column
bins_wdct = [-1, 90, 180, 270, 361]
labels_wdct = ['N', 'E', 'S', 'W']
data['wdct_class'] = pd.cut(data['wdct'], bins=bins_wdct, labels=labels_wdct)

# bin 'wdsp' column
bins_wdsp = [-9999, 5, 10, 15, 9999]
labels_wdsp = ['<5', '5 to 10', '10 to 15', '>15']
data['wdsp_class'] = pd.cut(data['wdsp'], bins=bins_wdsp, labels=labels_wdsp)
```

```python
# create transition matrix for 'temp_class' column
temp_transition_matrix = pd.crosstab(data['temp_class'].shift(), data['temp_class'], normalize='index').values

# create transition matrix for 'stp_class' column
stp_transition_matrix = pd.crosstab(data['stp_class'].shift(), data['stp_class'], normalize='index').values

# create transition matrix for 'hmdy_class' column
hmdy_transition_matrix = pd.crosstab(data['hmdy_class'].shift(), data['hmdy_class'], normalize='index').values

# create transition matrix for 'prcp_class' column
prcp_transition_matrix = pd.crosstab(data['prcp_class'].shift(), data['prcp_class'], normalize='index').values

# create transition matrix for 'wdct_class' column
wdct_transition_matrix = pd.crosstab(data['wdct_class'].shift(), data['wdct_class'], normalize='index').values

# create transition matrix for 'wdsp_class' column
wdsp_transition_matrix = pd.crosstab(data['wdsp_class'].shift(), data['wdsp_class'], normalize='index').values
```

```python
# set number of steps to predict
num_steps = 486  # number of rows from 1st Jan 2020 to 30th April 2021

# create an empty DataFrame to store the predicted values
predicted_df = pd.DataFrame(columns=['date', 'temp_class', 'stp_class', 'hmdy_class', 'prcp_class', 'wdct_class', 'wdsp_class'])

# predict values
for i in range(num_steps):
    # store the predicted values in a dictionary
    predicted_values = {
        'date': (pd.Timestamp('2020-01-01') + pd.Timedelta(days=i)).strftime('%Y-%m-%d'),
        'temp_class': initial_values['temp_class'],
        'stp_class': initial_values['stp_class'],
        'hmdy_class': initial_values['hmdy_class'],
        'prcp_class': initial_values['prcp_class'],
        'wdct_class': initial_values['wdct_class'],
        'wdsp_class': initial_values['wdsp_class']
    }
    # add the predicted values to the DataFrame
    predicted_df = pd.concat([predicted_df, pd.DataFrame(predicted_values, index=[0])], ignore_index=True)
    # update the initial values for the next step
    new_values = {}
    for variable in initial_values.keys():
        prev_class = initial_values[variable]
        tra_mat = transition_matrices[variable]
        row_index = var_label[variable].index(prev_class)
        transition_probs = tra_mat[row_index]
        new_class = str(np.random.choice(var_label[variable], p=transition_probs))
        if variable == 'prcp_class':
            print(new_class)
        new_values[variable] = new_class
    initial_values = new_values

# write the predicted values to a CSV file
predicted_df.to_csv('predicted_weather.csv', index=False)
```
