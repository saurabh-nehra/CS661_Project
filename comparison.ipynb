```python
import plotly.express as px
from plotly.subplots import make_subplots
import pandas as pd
```

```python
df = pd.read_csv('./weather_states1.csv')
predicted_df = pd.read_csv('./predicted.csv')
```

```python
df['date'] = df['date'].astype(str)
predicted_df['date'] = predicted_df['date'].astype(str)
```

```python
new_pdf = predicted_df.loc[predicted_df['date'] == '2020-01-01']
```

```python
new_df = df.loc[df['date'] == '2020-01-01']
```

```python
#Next-day weather (Actual vs Predicted)
fig_original = px.scatter_mapbox(data_frame=new_df,
                            lat="lat",
                            lon="lon",
                            color="stp_class",
                            color_discrete_map={"<980": "blue", "980 to 990": "green", "990 to 1000": "red", ">1000": "black"},
                            zoom=3,
                            mapbox_style="carto-positron",
                            )
fig_predicted = px.scatter_mapbox(data_frame=new_pdf,
                            lat="lat",
                            lon="lon",
                            color="stp_class",
                            color_discrete_map={"<980": "blue", "980 to 990": "green", "990 to 1000": "red", ">1000": "black"},
                            zoom=3,
                            mapbox_style="carto-positron",
                            )


fig_original.show()
fig_predicted.show()
```

```python
small_df = df[['date', 'inme', 'stp_class']]
small_pdf = predicted_df[['date', 'inme', 'stp_class']]
common_rows = pd.merge(small_df, small_pdf, on=['date', 'inme', 'stp_class'], how='inner')
```

```python
#Accuracy of model for 1 year (long term weather prediction of pressure)
print(common_rows.shape[0]/small_pdf.shape[0]*100)
```

```python
small_pdf1 = (predicted_df.groupby('inme')
                    .apply(lambda x: x.iloc[:1, :])
                    .reset_index(drop=True))
common_rows = pd.merge(small_df, small_pdf1, on=['date', 'inme', 'stp_class'], how='inner')
```

```python
#Accuracy of model for 1 day (next-day weather prediction accuracy of pressure)
print(common_rows.shape[0]/small_pdf1.shape[0]*100)
```
